```python
def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(x=10,y=20)
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(x=10,y="20")
```
